```r
setClass(
  "Individuo",
  slots = c(
    espacos = "numeric",
    valores = "numeric",
    limiteEspacos = "numeric",
    notaAvaliacao = "numeric",
    espacoUsado = "numeric",
    geracao = "numeric",
    cromossomo = "character"
  ),
  prototype = list(
    espacos = 0,
    valores = 0,
    limiteEspacos = 0,
    notaAvaliacao = 0,
    espacoUsado = 0,
    geracao = 0
      
  )
)
```

```r
gerar_cromossomo <- function(tamanho_cromossomo) {
    cromossomo <- sample(
        x = c("0", "1"), 
        size = tamanho_cromossomo, 
        replace = T
    ) 
    
    return(cromossomo)
}
```

### * Função de avaliação ( Fitness )
[ Avalia quanto o indivíduo está se ajustando ao problema ]

```r
# Medida de qualidade para saber como o cromossomo resolve o problema
# Se é uma solução aceitável e se pode ser utilizada para a evolução
```

```r
avaliacao <- function(individuo) {
    nota <- 0
    somaEspacos <- 0
    for (i in 1:length(individuo@cromossomo)) {
        if (individuo@cromossomo[i] == '1') {
            nota <- nota + individuo@valores[i]
            somaEspacos <- somaEspacos + individuo@espacos[i]
        }
    }
    
    if (somaEspacos > individuo@limiteEspacos) {
        nota = 1
    }
    
    individuo@notaAvaliacao <- nota
    individuo@espacoUsado <- somaEspacos
    
    return(individuo)
}
```

### * Crossover / reprodução
### Teoria ( Operador genético )
- Combina pedaços do cromossomo de dois genitores gerando filhos mais aptos e consequentemente com o passar das gerações a população tende a evoluir
- Representa a reprodução sexuada, pois na reprodução assexuada cada filho é idêntico a seu genitor e tem as mesmas habilidades, o que não cria diversidade

**Operador de crossover de um ponto**

Pai1 <- 1 0 1 1 0 1 0 (*7 genes: O número de elementos*)

Pai2 <- 0 1 0 0 1 1 1

Filho1 <- 1 0 1 0 1 1 1

Filho2 <- 0 1 0 1 0 1 0



### * Crossover / reprodução - Implementação

```r
# 2º Indivíduo
individuo2 <- new(
    "Individuo",
    espacos = espacos,
    valores = valores,
    limiteEspacos = limite
)
individuo2@cromossomo <- gerar_cromossomo(length(listaProdutos))
individuo2 <- avaliacao(individuo2)


cat(
    "\nEspaços: ", individuo2@espacos,
    "\nValores: ", individuo2@valores,
    "\nCromossomo: ", individuo2@cromossomo,
    "\nValor total R$", individuo2@notaAvaliacao,
    "\nEspaço total (m³): ", individuo2@espacoUsado
)
```

```r
# Função de crossover
crossover <- function(individuo_a, individuo_b) {
    indices <- 1:length(individuo_a@cromossomo)
    corte <- sample(indices, 1)
    
    cat("\nPonto de corte:", corte)
    
    if (corte == length(individuo_a@cromossomo)) {
        filho1 <- individuo_b@cromossomo[1:corte]
        filho2 <- individuo_a@cromossomo[1:corte]
    } else {
        filho1 <- c(individuo_b@cromossomo[1:corte], individuo_a@cromossomo[(corte+1):length(individuo_a@cromossomo)])
        filho2 <- c(individuo_a@cromossomo[1:corte], individuo_b@cromossomo[(corte+1):length(individuo_b@cromossomo)])
    }
    
    filhos <- list( # Construtor
        new(
            "Individuo", 
            espacos = individuo_a@espacos,
            valores = individuo_a@valores,
            limiteEspacos = individuo_a@limiteEspacos,
            geracao = individuo_a@geracao + 1,
            cromossomo = filho1
        ),
        new(
            "Individuo", 
            espacos = individuo_b@espacos,
            valores = individuo_b@valores,
            limiteEspacos = individuo_b@limiteEspacos,
            geracao = individuo_b@geracao + 1,
            cromossomo = filho2
        )
    )
    
  return(filhos)
    
}
```

```r
filhos <- crossover(
    individuo_a = individuo1,
    individuo_b = individuo2
)

cat(
    "\nIndivíduo 1:", individuo1@cromossomo,
    "\nIndivíduo 2:", individuo2@cromossomo,
    "\nFilho 1:", filhos[[1]]@cromossomo,
    "\nFilho 2:", filhos[[2]]@cromossomo
)
```

### * Mutação:
- A mutação cria diversidade, mudando aleatoriamente genes dentro de indivíduos é aplicada de forma menos frequente que a reprodução, como na natureza
*Gene representa cada célula do cromossomo* 

*Alelo é o valor dentro de cada gene*

- Possui uma taxa associada a uma probabilidade extremamente baixa para alterar um gene


```r
# Função de mutação
mutacao <- function(individuo, taxa_mutacao) {
    for (i in 1:length(individuo@cromossomo)) {
        if (runif(n = 1, min = 0, max = 1) < taxa_mutacao) {
            if (individuo@cromossomo[i] == '1') {
                individuo@cromossomo[i] <- '0'
            } else {
                individuo@cromossomo[i] <- '1'
            }
        }
    }
    
    return(individuo)
}
```

```r
cat("\nCromossomo antes da mutação:", individuo1@cromossomo)
individuo1 <- mutacao(individuo1, 0.05) # Onde terá uma probabilidade de 5% de ocorrer a mutação
cat("\nCromossomo depois da mutação:", individuo1@cromossomo)
```

### * Inicialização da população ( Gerar a população inicial )
- Um conjunto de indivíduos é chamado de **população**

```r
setClass(
    "algoritmoGenetico",
    slots = c( # Atributos da classe
        tamanhoPopulacao = "numeric",
        populacao = "list",
        geracao = "numeric",
        melhorSolucao = "Individuo"
    )
)
```

```r
# Função de inicialização da população
inicializa_populacao <- function(algoritmo_genetico, espacos, valores, limite) {
    for (i in 1:algoritmo_genetico@tamanhoPopulacao) {
        algoritmo_genetico@populacao[[i]] <- 
        new(
            "Individuo", 
            espacos = espacos,
            valores = valores,
            limiteEspacos = limite
        )
        
        algoritmo_genetico@populacao[[i]]@cromossomo <- gerar_cromossomo(length(espacos))
        
    }
    
    return(algoritmo_genetico)
}
```

```r
espacos <- c()
valores <- c()
nomes <- c()

for (produto in listaProdutos) {
    espacos <- c(espacos, produto@espaco)
    valores <- c(valores, produto@valor)
    nomes <- c(nomes, produto@nome)
}

limite <- 3
tamanho <- 20

ag <- new(
    "algoritmoGenetico",
    tamanhoPopulacao = tamanho
)
ag <- inicializa_populacao(
    algoritmo_genetico = ag,
    espacos = espacos,
    valores = valores,
    limite = limite
)
```

```r
for (i in 1:ag@tamanhoPopulacao) {
    cat(
        "\n ***** Indivíduo", i, "*****",
        "\nEspaços:", ag@populacao[[i]]@espacos,
        "\nValores:", ag@populacao[[i]]@valores,
        "\nCromossomo:", ag@populacao[[i]]@cromossomo
    )
}
```

### * Avaliação da população

*Priorizando indivíduos com as maiores notas*


```r
# Função de ordenamento da população
ordena_populacao <- function(populacao) {
    populacao_ordenada <- c()
    notas_avaliacao <- c()
    for (individuo in populacao) {
        notas_avaliacao <- c(notas_avaliacao, individuo@notaAvaliacao)
    }
    lista_posicao <- order(notas_avaliacao, decreasing = T)
    for (i in 1:length(lista_posicao)) {
        populacao_ordenada <- c(populacao_ordenada, populacao[[lista_posicao[i]]])
    }
    
    return(populacao_ordenada)
}
```

```r
espacos <- c()
valores <- c()
nomes <- c()

for (produto in listaProdutos) {
    espacos <- c(espacos, produto@espaco)
    valores <- c(valores, produto@valor)
    nomes <- c(nomes, produto@nome)
}

limite <- 3
tamanho <- 20

ag <- new(
    "algoritmoGenetico",
    tamanhoPopulacao = tamanho
)
ag <- inicializa_populacao(
    algoritmo_genetico = ag,
    espacos = espacos,
    valores = valores,
    limite = limite
)

for (i in 1:ag@tamanhoPopulacao) {
    ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
}

ag@populacao <- ordena_populacao(ag@populacao)

for (i in 1:ag@tamanhoPopulacao) {
    cat(
        "\n ***** Indivíduo", i, "*****",
        "\nEspaços:", ag@populacao[[i]]@espacos,
        "\nValores:", ag@populacao[[i]]@valores,
        "\nCromossomo:", ag@populacao[[i]]@cromossomo,
        "\nValor total R$", ag@populacao[[i]]@notaAvaliacao
    )
}
```

### * Melhor indivíduo

```r
# Função responsável por recuperar o melhor indivíduo
melhor_individuo <- function(algoritmo_genetico, individuo) {
    if (individuo@notaAvaliacao > algoritmo_genetico@melhorSolucao@notaAvaliacao) {
        algoritmo_genetico@melhorSolucao <- individuo
    }
    
    return(algoritmo_genetico)
}
```

```r
espacos <- c()
valores <- c()
nomes <- c()

for (produto in listaProdutos) {
    espacos <- c(espacos, produto@espaco)
    valores <- c(valores, produto@valor)
    nomes <- c(nomes, produto@nome)
}

limite <- 3
tamanho <- 20

ag <- new(
    "algoritmoGenetico",
    tamanhoPopulacao = tamanho
)
ag <- inicializa_populacao(
    algoritmo_genetico = ag,
    espacos = espacos,
    valores = valores,
    limite = limite
)

for (i in 1:ag@tamanhoPopulacao) {
    ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
}

# Inicializando o atributo com algum valor
ag@melhorSolucao <- ag@populacao[[1]]
ag@populacao <- ordena_populacao(ag@populacao)

ag <- melhor_individuo(
    algoritmo_genetico = ag, 
    individuo = ag@populacao[[1]] # Atribuindo o primeiro elemento, depois da ordenação, onde o primeiro indivíduo será o melhor de todos
)

cat(
    "\n**** Melhor solução ***", 
    "\nCromossomo:", ag@melhorSolucao@cromossomo,
    "\nEspaço total:", ag@melhorSolucao@espacoUsado,
    "\nValor total R$", ag@melhorSolucao@notaAvaliacao
)
```

### * Soma das avaliações

```r
soma_avaliacoes <- function(algoritmo_genetico) {
    soma <- 0
    for (individuo in algoritmo_genetico@populacao) {
        soma = soma + individuo@notaAvaliacao
    }
    
    return(soma)
}
```

```r
espacos <- c()
valores <- c()
nomes <- c()

for (produto in listaProdutos) {
    espacos <- c(espacos, produto@espaco)
    valores <- c(valores, produto@valor)
    nomes <- c(nomes, produto@nome)
}

limite <- 3
tamanho <- 20

ag <- new(
    "algoritmoGenetico",
    tamanhoPopulacao = tamanho
)
ag <- inicializa_populacao(
    algoritmo_genetico = ag,
    espacos = espacos,
    valores = valores,
    limite = limite
)

for (i in 1:ag@tamanhoPopulacao) {
    ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
}

# Inicializando o atributo com algum valor
ag@melhorSolucao <- ag@populacao[[1]]
ag@populacao <- ordena_populacao(ag@populacao)

ag <- melhor_individuo(
    algoritmo_genetico = ag, 
    individuo = ag@populacao[[1]] # Atribuindo o primeiro elemento, depois da ordenação, onde o primeiro indivíduo será o melhor de todos
)

cat(
    "\n**** Melhor solução ***", 
    "\nCromossomo:", ag@melhorSolucao@cromossomo,
    "\nEspaço total:", ag@melhorSolucao@espacoUsado,
    "\nValor total R$", ag@melhorSolucao@notaAvaliacao
)


soma <- soma_avaliacoes(algoritmo_genetico = ag)
cat("\nSoma das avaliações:", soma)
```

### * Seleção dos indivíduos - Teoria



- Operadores genéticos são utilizados em indivíduos selecionados dentro da população

- Indivíduos mais aptos (Função de avaliação maior) serão selecionados mais frequentemente que os menos aptos - Características dos melhores deve predominar na nova população

- Deve simular o mecanismo de seleção natural que atua sobre as espécies biológicas: Pais mais capazes geram mais filhos e pais menos aptos também geram descendentes

- Privilegiar indivíduos com função de avaliação alta, sem desprezar completamente com função baixa

- Se deixar somente os melhores indivíduos, a população tenderá a ser composta de elementos cada vez mais semelhantes e faltará diversidade

### * Seleção dos indivíduos: Método da roleta viciada

- Cada cromossomo recebe um pedaço proporcional à sua avaliação e a roleta é rodada

- Etilismo: Módulos de população que preservam os melhores, que garantem a estabilidade do melhor e não sua evolução



### * Seleção dos indivíduos - Implementação

```r
seleciona_pai <- function(algoritmo_genetico, soma_avaliacoes) {
    pai <- 0
    valor_sorteado <- runif(1, min = 0, max = 1) * soma_avaliacoes
    soma <- 0
    i <- 1
    while (i < length(algoritmo_genetico@populacao) & soma < valor_sorteado) {
        soma <- soma + algoritmo_genetico@populacao[[i]]@notaAvaliacao
        pai <- pai + 1
        i <- i + 1
    }
    
    return(pai)
}
```

```r
espacos <- c()
valores <- c()
nomes <- c()

for (produto in listaProdutos) {
    espacos <- c(espacos, produto@espaco)
    valores <- c(valores, produto@valor)
    nomes <- c(nomes, produto@nome)
}

limite <- 3
tamanho <- 20

ag <- new(
    "algoritmoGenetico",
    tamanhoPopulacao = tamanho
)
ag <- inicializa_populacao(
    algoritmo_genetico = ag,
    espacos = espacos,
    valores = valores,
    limite = limite
)

for (i in 1:ag@tamanhoPopulacao) {
    ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
}

# Inicializando o atributo com algum valor
ag@melhorSolucao <- ag@populacao[[1]]
ag@populacao <- ordena_populacao(ag@populacao)

ag <- melhor_individuo(
    algoritmo_genetico = ag, 
    individuo = ag@populacao[[1]] # Atribuindo o primeiro elemento, depois da ordenação, onde o primeiro indivíduo será o melhor de todos
)

soma <- soma_avaliacoes(
    algoritmo_genetico = ag
)

for (individuos_gerados in 1:(ag@tamanhoPopulacao / 2)) {
    pa1 <- seleciona_pai(
        algoritmo_genetico = ag, 
        soma_avaliacoes = soma
    )
    
    pai2 <- seleciona_pai(
        algoritmo_genetico = ag, 
        soma_avaliacoes = soma
    )
}
```

### * Construção de nova geração

```r
espacos <- c()
valores <- c()
nomes <- c()

for (produto in listaProdutos) {
    espacos <- c(espacos, produto@espaco)
    valores <- c(valores, produto@valor)
    nomes <- c(nomes, produto@nome)
}

limite <- 3
tamanho <- 20

ag <- new(
    "algoritmoGenetico",
    tamanhoPopulacao = tamanho
)

# 1º rodada -------------------------------

ag <- inicializa_populacao(
    algoritmo_genetico = ag,
    espacos = espacos,
    valores = valores,
    limite = limite
)

for (i in 1:ag@tamanhoPopulacao) {
    ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
}

# Inicializando o atributo com algum valor
ag@melhorSolucao <- ag@populacao[[1]]
ag@populacao <- ordena_populacao(ag@populacao)

ag <- melhor_individuo(
    algoritmo_genetico = ag, 
    individuo = ag@populacao[[1]] # Atribuindo o primeiro elemento, depois da ordenação, onde o primeiro indivíduo será o melhor de todos
)

soma <- soma_avaliacoes(
    algoritmo_genetico = ag
)

# 2º rodada --------------------------------

nova_populacao <- c()
probabilidade_mutacao <- 0.05


for (individuos_gerados in 1:(ag@tamanhoPopulacao / 2)) {
    
    pai1 <- 1
    pai2 <- 1
    
    while (pai1 == pai2) {
      pai1   <- seleciona_pai(
        algoritmo_genetico = ag, 
        soma_avaliacoes = soma
      )

      pai2 <- seleciona_pai(
        algoritmo_genetico = ag, 
        soma_avaliacoes = soma
      )
    }
   
    cat("\nÍndice Pai 1:", pai1," - Índice Pai 2:", pai2)
    
    
    filhos <- crossover(
        individuo_a = ag@populacao[[pai1]],
        individuo_b = ag@populacao[[pai2]]
    )
    nova_populacao <- c(
        nova_populacao, 
        mutacao(
            individuo = filhos[[1]],
            taxa_mutacao = probabilidade_mutacao
        )
    )
    nova_populacao <- c(
        nova_populacao, 
        mutacao(
            individuo = filhos[[2]],
            taxa_mutacao = probabilidade_mutacao
        )
    )
    
}

ag@populacao <- nova_populacao
for (i in 1:ag@tamanhoPopulacao) {
    ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
}

ag@populacao <- ordena_populacao(ag@populacao)

ag <- melhor_individuo(
    algoritmo_genetico = ag, 
    individuo = ag@populacao[[1]] # Atribuindo o primeiro elemento, depois da ordenação, onde o primeiro indivíduo será o melhor de todos
)

soma <- soma_avaliacoes(
    algoritmo_genetico = ag
)
```

### * Algoritmo genético completo

```r
visualiza_geracao <- function(algoritmo_genetico) {
    melhor <- algoritmo_genetico@populacao[[1]]
    cat(
        "\nG:", melhor@geracao, 
        "Valor total:", melhor@notaAvaliacao,
        "Espaço total:", melhor@espacoUsado,
        "Cromossomo:", melhor@Cromossomo
        
    )
}
```

```r
resolver <- function(algoritmo_genetico, taxa_mutacao, num_geracoes, espacos, valores, limite_espaco) {
    ag <- algoritmo_genetico
    ag <- inicializa_populacao(
        algoritmo_genetico = ag,
        espacos = espacos,
        valores = valores,
        limite = limite_espaco
    )
    
    for (i in 1:ag@tamanhoPopulacao) {
        ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
    }
    
    ag@populacao <- ordena_populacao(ag@populacao)
    ag@melhorSolucao <- ag@populacao[[1]]
    visualiza_geracao(algoritmo_genetico = ag) 
    
    for (i in 1:num_geracoes) {
        soma = soma_avaliacoes(algoritmo_genetico = ag)
        nova_populacao <- c()
        for(individuos_gerados in 1:(ag@tamanhoPopulacao / 2)) {
            
            pai1 <- 1
            pai2 <- 1
            
           while (pai1 == pai2) {
              pai1   <- seleciona_pai(
                algoritmo_genetico = ag, 
                soma_avaliacoes = soma
              )

              pai2 <- seleciona_pai(
                algoritmo_genetico = ag, 
                soma_avaliacoes = soma
              )
            }
            
          # Combinando os pais e criando os dois filhos
          filhos = crossover(
              individuo_a = ag@populacao[[pai1]],
              individuo_b = ag@populacao[[pai2]]
          )
           
            filho1 <- mutacao(
                individuo = filhos[[1]], 
                taxa_mutacao = 0.05
            )
            
            filho2 <- mutacao(
                individuo = filhos[[2]], 
                taxa_mutacao = 0.05
            )
            
            filhos_gerados <- c(filho1, filho2)
            
            nova_populacao <- c(nova_populacao, filhos_gerados)   
        }
        
        ag@populacao <- nova_populacao
        # Realizando a avaliação da nova população
        for (i in 1:ag@tamanhoPopulacao) {
            ag@populacao[[i]] <- avaliacao(ag@populacao[[i]])
        }
        ag@populacao <- ordena_populacao(ag@populacao)
        visualiza_geracao(algoritmo_genetico = ag) 
        ag <- melhor_individuo(
            algoritmo_genetico = ag,
            individuo = ag@populacao[[1]]
        )
    }
    
    
   cat(
    "\n**** Melhor solução ***", 
    "\nGeração:", ag@melhorSolucao@geracao,
    "\nCromossomo:", ag@melhorSolucao@cromossomo,
    "\nEspaço total:", ag@melhorSolucao@espacoUsado,
    "\nValor total R$", ag@melhorSolucao@notaAvaliacao
    ) 
    
    return(ag)
    
}
```
